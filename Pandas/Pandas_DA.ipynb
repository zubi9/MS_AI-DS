```python
# Loading Lib
import pandas as pd
```

```python
# Loading Data Note: Important to spacify the sheet you are working on
data1 = pd.read_excel('/content/test_data.xlsx', sheet_name = 'Year 2009-2010')
```

```python
# slicing our desired Data
data1 = data1.iloc[433200: 437200]
data1
```

```python
# we must have information about the data that we are working with
data1.info()
```

```python
data1.describe()
```

```python
# Droping Null Values in axis=0 means droping whole Row which is containg it
df = data1.dropna(axis=0)
df
```

```python
# Question 1.
len(df.Invoice.unique())
```

```python
# max(df.Country)
```

```python
# min(df.Country)
```

```python
# created new col with Total sale of one item
df['Total'] = df.Quantity * df.Price
```

```python
# Question 2
max_cn = df[(df.Country == max(df.Country))]['Total'].count()
print(f"Country {max(df.Country)} with maximum orders of : {max_cn}")
```

```python
# Question 3
min_cn = df[(df.Country == min(df.Country))]['Total'].count()
print(f"Country {min(df.Country)} with minimum orders of : {min_cn}")
```

```python
# Question 4.
df.groupby(df.Invoice)['Total'].sum().nlargest(1)#.sort_values(ascending=False)
```

```python
# Question 5
# find Id of the
orders_sum = df.groupby(df['Customer ID'])['Total'].sum().sort_values(ascending=False).nlargest(3)
print("ids wiht aorders amount\n", orders_sum)
norders = df[df['Customer ID'] == 12477.0].Invoice.nunique()
print("\nNo of the Times the seconed Highest spending customer ordered = ", norders)
```
