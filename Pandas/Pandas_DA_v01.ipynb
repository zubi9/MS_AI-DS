```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd

data = pd.read_excel('/content/test_data.xlsx', sheet_name = 'Year 2009-2010')
data = data.iloc[433200 : 437200]
selected_data  = data.dropna(axis=0)
```

```python

# Answering the questions:

# 1. The amount of unique invoices
unique_invoices = selected_data['Invoice'].nunique()
selected_data['Total'] = selected_data.Quantity * selected_data.Price
# 2. The country with the most orders
most_orders_country = selected_data['Country'].value_counts().idxmax()
most_orders_count = selected_data['Country'].value_counts().max()

# 3. The country with the least orders
least_orders_country = selected_data['Country'].value_counts().idxmin()
least_orders_count = selected_data['Country'].value_counts().min()

# 4. The highest total price per item in one order
max_price_per_item_order = selected_data.groupby('Invoice')['Total'].sum().max()
max_price_per_item_order_id = selected_data.groupby('Invoice')['Total'].sum().idxmax()

# 5. The second highest spending customer
top_customers = selected_data.groupby('Customer ID')['Total'].sum().sort_values(ascending=False)

# 5. The second highest spending customer and the total number of times they ordered
second_highest_spending_customer = top_customers.index[1]
second_highest_spending_customer_orders = selected_data[selected_data['Customer ID'] == second_highest_spending_customer]['Invoice'].nunique()

# Displaying the updated result

print(f"1. The amount of unique invoices is {unique_invoices}.")
print(f"2. The country with the most orders is {most_orders_country} with {most_orders_count} orders.")
print(f"3. The country with the least orders is {least_orders_country} with {least_orders_count} orders.")
print(f"4. The highest total price per item in one order was {max_price_per_item_order:.2f} in order with id {max_price_per_item_order_id}.")
print(f"5. The second highest spending customer has id {second_highest_spending_customer} and ordered {second_highest_spending_customer_orders} times.")
```
